<h1> Preprocessing the image

<h3> importing python libraries

```python
import cv2
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import scipy.fftpack
import theano
from theano import tensor as T
from theano.tensor.nnet import conv2d
```

<h3> plots the images

```python
def showImage(img,gimg,pimg,title):
    plt.subplot(131),plt.imshow(img),plt.title('Original')
    plt.xticks([]), plt.yticks([])
    plt.subplot(132),plt.imshow(gimg),plt.title('Gratscale')
    plt.xticks([]), plt.yticks([])
    plt.subplot(133),plt.imshow(pimg),plt.title(title)
    plt.xticks([]), plt.yticks([])
    plt.show()
```

<h3> resizes the image

```python
def resize(img,w,h):
    res = cv2.resize(img, (w, h), interpolation = cv2.INTER_CUBIC)
    return res
```

<h3> loads an image and converts to grayscale

```python
img = cv2.imread('/home/dirie/HonoursProject/IPYTHON/data/00002.jpg')
gimg = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

<h5> bluring the image using median filter

```python
img1 = cv2.medianBlur(gimg,5)
showImage(img,gimg,img1,"Median-blur")
```

<h5> threshold the image 

```python
(thresh, im_bw) = cv2.threshold(img1, 127, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
#showImage(img,gimg,im_bw,"threshold")
```

<h5> adabtive threshold the image 

```python
thresh = cv2.adaptiveThreshold(im_bw,255,1,1,11,2)
#showImage(img,gimg,thresh,"Adaptive-Threshold")
```

<h5> dilation the image 

```python
thresh = cv2.dilate(thresh,None,iterations = 2)
#showImage(img,gimg,thresh,"dilation")
```

<h5> eroding the image 

```python
thresh = cv2.erode(thresh,None,iterations = 2)
#showImage(img,gimg,thresh,"eroding")
```

<h5> resizng the image 

```python
rimg = resize(thresh,10,10)
#showImage(img,gimg,thresh,"resized")
```
